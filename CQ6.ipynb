except: en español su significado es excepto, este solo se ejecutará cuando el bloque try falle
debido a un error. Este bloque contiene sentencias que generalmente nos dan un contexto de lo
que salió mal en el bloque try.

# Esto se ejecutará si el bloque try se ejecuta sin errores

# Este bloque se ejecutará siempre

Finally: en español significaría finalmente, esta siempre se ejecuta después de los bloques o
funciones de try y except. Siempre se ejecuta después de la ejecución normal del bloque de
prueba o después de que finaliza el bloque de prueba debido a alguna excepción

with: la función de delimitar el espacio de nombre(“namespaces”) como se hacen en otros
lenguajes como VB. Es más para determinar la configuración local que tendrá un bloque de código,
lo que se conoce como “contexto”.

para recuperar los valores anteriores. Un ejemplo sería la apertura de un fichero:
